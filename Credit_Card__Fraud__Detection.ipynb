```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
train_df=pd.read_csv("fraudTrain.csv")
train_df.head()
```

```python
test_df=pd.read_csv("fraudTest.csv")
test_df.head()
```

```python
combined_df=pd.concat([train_df, test_df], ignore_index=True)
data=combined_df
```

```python
good = data [data.is_fraud == 0]
fraud = data [data.is_fraud == 1]
```

```python
good_sample = good.sample(n=25)
```

```python
new_data = pd.concat([good_sample,fraud],axis=0)
```

```python
new_data.tail()
```

```python
x1 = new_data.drop('is_fraud',axis = 1)
```

```python
y1 = new_data['is_fraud']
```

```python
y1
```

```python
x1
```

```python
x1 = x1.apply(pd.to_numeric, errors='coerce')
y1 = y1.apply(pd.to_numeric, errors='coerce')
```

```python
x1.fillna(0, inplace=True)
y1.fillna(0, inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x1,y1,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
x1_prediction = log.predict(X_train)
```

```python
x1_accuracy = accuracy_score(x1_prediction,y_train)
```

```python
x1_accuracy
```

```python
y1_predict = log.predict(X_test)
```

```python
y1_accuracy = accuracy_score(y1_predict,y_test)
```

```python
y1_accuracy
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
x1_precision = precision_score(x1_prediction,y_train)
```

```python
y1_precision = precision_score(y1_predict,y_test)
```
